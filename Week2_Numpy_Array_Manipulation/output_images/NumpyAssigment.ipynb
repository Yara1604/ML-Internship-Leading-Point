```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
from PIL import Image
```

```python
shape = (1000, 1000, 3)
```

```python
# Red images

img1 = Image.open('C:/Users/yarah/Projects/LeadingPoint/sea_image.jpg')
img1 = img1.convert('RGB')
img1 = img1.resize((1000, 1000)) # resizes image in-place
red_arr = np.array(img1)

# np.zeros(shape, dtype=np.uint8)
# red_arr[:, :, :] = [255, 0, 0]
```

```python
# Green images
img2 = Image.open('C:/Users/yarah/Projects/LeadingPoint/flower_image.jpg')

img2 = img2.convert('RGB')

img2 = img2.resize((1000, 1000))

green_arr = np.array(img2)

# np.zeros(shape, dtype=np.uint8)

# green_arr[:, :, :] = [0, 255, 0]
```

```python
# output 1

output1 = np.vstack((red_arr[:500,:1000],green_arr[500:1000,:]))
plt.imshow(output1)
plt.show
```

```python
# output 2
output2 = np.hstack((red_arr[:1000,:500],green_arr[:1000,500:1000]))
plt.imshow(output2)
plt.show()
```

```python
# output 3
row_1 = np.hstack((red_arr[0:500,0:500],green_arr[0:500,500:1000]))
row_2 = np.hstack((green_arr[500:1000,0:500], red_arr[500:1000, 500:1000]))

output3 = np.vstack((row_1, row_2))
plt.imshow(output3)
plt.show()
```

```python
# output 4
# 5 red / green striped from top to bottom, 1000/10 = 100

output4 = red_arr.copy()

output4[100:200,:] = green_arr[100:200,:]
output4[300:400,:] = green_arr[300:400,:]
output4[500:600,:] = green_arr[500:600,:]
output4[700:800,:] = green_arr[700:800,:]
output4[900:1000,:] = green_arr[900:1000,:]


plt.imshow(output4)
plt.show()
```

```python
# output 5

# 5 red / green striped from top to bottom, 1000/10 = 100
output5 = red_arr.copy()

output5[:,100:200] = green_arr[:,100:200]
output5[:,300:400] = green_arr[:,300:400]
output5[:,500:600] = green_arr[:,500:600]
output5[:,700:800] = green_arr[:,700:800]
output5[:,900:1000] = green_arr[:,900:1000]


plt.imshow(output5)
plt.show()
```

```python
# output 6

output6 = red_arr.copy() 

for i in range(0,1000,100):
    # If the row number is even, start c at 0. 
    # If the row number is odd, start c at 1 to offset the pattern.

    # if i is 100, that means row 1
    if ((i / 100) % 2 == 0):
        c = 0
    else:
        c = 1

    for j in range(0,1000,100):
        if c%2 == 0:
            output6[i:i+100,j:j+100] = green_arr[i:i+100,j:j+100]
        c+=1

plt.imshow(output6)
plt.show()
plt.imsave('MyOutput6.png', output6)
```
